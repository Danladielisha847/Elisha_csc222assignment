```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#For Pie chart
# 1. Proportion of Depression Cases

plt.figure(figsize=(6, 6))
shima['Depression'].value_counts().plot.pie(autopct='%1.1f%%', labels=['No', 'Yes'])
plt.title('Proportion of Depression Cases')
plt.ylabel('')
plt.show()
```

```python
# 2. Gender Distribution

plt.figure(figsize=(6, 6))
shima['Gender'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
# 3. Sleep Duration Categories

plt.figure(figsize=(6, 6))
shima['Sleep Duration'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Sleep Duration Distribution')
plt.ylabel('')
plt.show()
```

```python
# 4. Suicidal Thoughts (Yes/No)

plt.figure(figsize=(6, 6))
shima['Have you ever had suicidal thoughts ?'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Proportion of Suicidal Thoughts')
plt.ylabel('')
plt.show()
```

```python
# FOR BAR CHART
```

```python
# 1. Depression by Gender (bar chart)

plt.figure(figsize=(6, 4))
sns.countplot(data=shima, x='Gender', hue='Gender', legend= False, palette='viridis')
plt.title('Depression Cases by Gender')
plt.show()
```

```python
# 2. Academic Pressure Distribution(bar chart)

plt.figure(figsize=(8, 4))
sns.countplot(data=shima, x='Academic Pressure', color='skyblue', hue='Academic Pressure', legend= False, palette='muted')
plt.title('Distribution of Academic Pressure Levels (1-5)')
plt.show()
```

```python
#3. Sleep Duration vs. Depression(bar chart)

plt.figure(figsize=(8, 4))
sns.countplot(data=shima, x='Sleep Duration', hue='Sleep Duration', legend= False, palette='rocket')
plt.title('Sleep Duration vs. Depression')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Suicidal Thoughts by Age Group

shima['Age Group'] = pd.cut(shima['Age'], bins=[15, 20, 25, 30, 35, 40])
plt.figure(figsize=(8, 4))
sns.countplot(data=shima, x='Age Group', hue='Have you ever had suicidal thoughts ?')
plt.title('Suicidal Thoughts by Age Group')
plt.show()
```
